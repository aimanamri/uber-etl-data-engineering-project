## Concatenate all Parquet files by years

```python
import glob
import pandas as pd
```

```python
def create_df_year(year):
    dfs = []
    # Read parquet files from directory /data/{year} 
    for fin in glob.glob(f'./data/{year}/*.parquet'):
      dfs.append(pd.read_parquet(fin))
    df = pd.concat(dfs)
    return df

def save_as_parquet(df,name = 'df'):
        return df.to_parquet(f'{name}.parquet.gzip',
              compression='gzip')
```

##  Get the first N rows randomly from each month
`*.parquet.gzip` files are still too large to upload into Github and take too much storage/bandwidth even with Git LFS. (about 1.5GB)

Hence, reducing the file size is the best solution by filtering the rows for this side project use. Here, we will use first `20,000` rows randomly from each month

### 2022 Dataset CSV

```python
df = pd.read_parquet('2022.parquet.gzip').dropna(subset=['passenger_count', 'RatecodeID'], how='all')
df = df.groupby([df['tpep_pickup_datetime'].dt.month]).apply(lambda x: x.sample(20000)).reset_index(drop=True).sort_values(by='tpep_pickup_datetime')
df
```

```python
# Check the count of rows if 10,000 rows were exactly selected or not.
df['tpep_pickup_datetime'].groupby([df['tpep_pickup_datetime'].dt.month]).agg('count')
```

```python
df.to_csv('data_2022.csv', index=False)
```
